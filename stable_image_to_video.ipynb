```python
import torch
from diffusers import StableVideoDiffusionPipeline
from diffusers.utils import load_image, export_to_video

# Load the conditioning image
image = load_image("/content/7.png")
image = image.resize((576, 576))

# Initialize the pipeline
pipe = StableVideoDiffusionPipeline.from_pretrained(
    "stabilityai/stable-video-diffusion-img2vid-xt", torch_dtype=torch.float16, variant="fp16"
)
pipe.enable_model_cpu_offload()

# Set random seed for reproducibility
generator = torch.manual_seed(42)

# Generate frames using the pipeline
frames = pipe(image, decode_chunk_size=8, generator=generator).frames[0]

# Resize frames to square aspect ratio (e.g., 576x576)
square_frames = [frame.resize((576, 576)) for frame in frames]

# Export the square frames to video
export_to_video(square_frames, "generated_square.mp4", fps=7)
```
