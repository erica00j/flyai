```python
import cv2, sys

inVidName = '/content/video-output1.mp4'
outVidName = 'output.mp4'

cap = cv2.VideoCapture(inVidName)

if (cap.isOpened()==False):
  print("Error Opening")
  sys.exit()

tS = 0.5

tTxt = 0.25

txt = "My friend is crying "

fc = (0,0,0)

fs = 1

fo = cv2.FONT_HERSHEY_COMPLEX_SMALL

ft = 1

xW = 0.10
xH = 0.80

FW = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
FH = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
x = int(FW*xW)
y = int(FH*xH)

fps = int(cap.get(cv2.CAP_PROP_FPS))
nframe = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
duration = nframe/fps
minutes = int(duration/60)
seconds = duration%60

nT = len(txt)
txtWidth, _ = cv2.getTextSize(txt, fo, fs, ft)
if (x+txtWidth[0])>=FW:
  print("\n All text cannot be printed")
  print("\n Width of the video: ", FW)
  print("\n Width of the text excluding start location: ",txtWidth[0])
  print("\n Excess width in pixels: ", x + txtWidth[0] - FW,"\n")
  sys.exit()

tTxtDura = nT * tTxt
if (tS + tTxtDura) > duration:
  print("\n Text cannot be added before end of video")
  sys.exit()

nS = int(tS * fps)
nE = int(nS + nT * (fps*tTxt))

codec = cv2.VideoWriter_fourcc(*'mp4v')
video = cv2.VideoWriter(outVidName, codec, fps, (FW,FH))

i = 1
j = 0
while (cap.isOpened()==True):
  ret, frm = cap.read()
  if (ret == True):
    if (i<nS):
      video.write(frm)
    if (i>nS and i < nE):
      m = int((i - nS)/(fps*tTxt))
      j = m
      if (j<=nT):
        for k in range(j+1):
          cv2.putText(frm,txt[0:k],(x,y),fo,fs,fc,ft)
        j = j+1
        video.write(frm)
    if (i > nE):
      cv2.putText(frm, txt[0:k], (x, y), fo, fs, fc, ft)
      video.write(frm)
    i = i+1
  else:
    break
cap.release()
video.release()
cv2.destroyAllWindows()
```

```python
import cv2
import sys

inVidName = '/content/다운로드 (1).mp4'
outVidName = 'output3.mp4'

cap = cv2.VideoCapture(inVidName)

if not cap.isOpened():
    print("Error Opening")
    sys.exit()

tS = 0.5
tTxt = 0.25
txt = "My friend is crying "
bg_color = (0, 0, 0)  # Black background color
text_color = (255, 255, 255)  # White text color
fs = 1
fo = cv2.FONT_HERSHEY_COMPLEX_SMALL
ft = 1
xW = 0.10
xH = 0.80

FW = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
FH = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
x = int(FW * xW)
y = int(FH * xH)

fps = int(cap.get(cv2.CAP_PROP_FPS))
nframe = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
duration = nframe / fps
minutes = int(duration / 60)
seconds = duration % 60

nT = len(txt)
txtWidth, _ = cv2.getTextSize(txt, fo, fs, ft)
if (x + txtWidth[0]) >= FW:
    print("\n All text cannot be printed")
    print("\n Width of the video: ", FW)
    print("\n Width of the text excluding start location: ", txtWidth[0])
    print("\n Excess width in pixels: ", x + txtWidth[0] - FW, "\n")
    sys.exit()

tTxtDura = nT * tTxt
if (tS + tTxtDura) > duration:
    print("\n Text cannot be added before the end of the video")
    sys.exit()

nS = int(tS * fps)
nE = int(nS + nT * (fps * tTxt))

codec = cv2.VideoWriter_fourcc(*'mp4v')
video = cv2.VideoWriter(outVidName, codec, fps, (FW, FH))

i = 1
j = 0
while cap.isOpened():
    ret, frm = cap.read()
    if ret:
        if i < nS:
            video.write(frm)
        if nS < i < nE:
            m = int((i - nS) / (fps * tTxt))
            j = m
            if j <= nT:
                # Add a black background rectangle behind the text
                cv2.rectangle(frm, (x, y - 20), (x + txtWidth[0], y + 5), bg_color, thickness=cv2.FILLED)
                # Draw the text in white color
                for k in range(j + 1):
                    cv2.putText(frm, txt[0:k], (x, y), fo, fs, text_color, ft)
                j = j + 1
                video.write(frm)
        if i > nE:
            # Add a black background rectangle behind the text
            cv2.rectangle(frm, (x, y - 20), (x + txtWidth[0], y + 5), bg_color, thickness=cv2.FILLED)
            # Draw the text in white color
            cv2.putText(frm, txt[0:k], (x, y), fo, fs, text_color, ft)
            video.write(frm)
        i = i + 1
    else:
        break

cap.release()
video.release()
cv2.destroyAllWindows()
```
